# Rosenbrock function

Simple example and plot of Rosenbrock function in Python, with sample line.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
%matplotlib inline
```

```python
def rosenbrock(X, Y):
    return (1-X)**2 + 100*(Y-X**2)**2
```

```python
x = np.arange( -3, 3, 0.05)
y = np.arange( -6, 6, 0.05)
x, y = np.meshgrid(x,y)
```

```python
path_y = np.arange(-6,6,0.05)
path_x = np.zeros(path_y.size)
path_x, path_y = np.meshgrid(path_x, path_y)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(y,x,rosenbrock(x,y), cmap=cm.coolwarm, linewidth=0)
line = ax.plot(path_y, path_x, rosenbrock(path_x,path_y)[:,1], '--k')
point = ax.plot()
ax.hold(True)
```
